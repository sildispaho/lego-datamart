```python
query = """
SELECT
    dc.Name AS Category,
    dco.Name AS Color,
    fp.Nr_of_Parts
FROM Fact_Parts fp
JOIN Dim_Category dc ON fp.Category_ID = dc.ID
JOIN Dim_Color dco ON fp.Color_ID = dco.ID
"""
df = pd.read_sql(query, conn)
df.head()
```
